### Specify the wealth array $W$

```julia
number_of_steps = 100;
W = range(1,10.0, length=number_of_steps) |> collect;
```

## Bernoulli Utility function $U(u) = \ln(w)$
The utility function used in the [St. Petersburg]() paradox is the natural log of the decions makers wealth. This utility function describes only risk adverse decision makers. To see this, let's compute the risk aversion measure $r(w)$. The first and second derivatives of the utility function are given by:

$$
\begin{eqnarray}
U^{\prime}(w) & = & \frac{1}{w} \\
U^{\prime\prime}(w) & = & -\frac{1}{w^{2}} \\
\end{eqnarray}
$$

which gives the risk measure $r(w)$ of

$$
\begin{equation}
r(w) = \frac{1}{w}
\end{equation}
$$

which is always positive.

### Simulation

```julia
let
    
    R(w) = 1/w;
    
    simulation_array = Array{Float64,2}(undef, number_of_steps, 2)
    for i  eachindex(W)
        simulation_array[i,1] = W[i]
        simulation_array[i,2] = R(W[i])
    end
    
    plot(simulation_array[:,1], simulation_array[:,2], lw=3, label="U(w) = ln(w)", c=:navyblue,
        xlabel="Wealth w", ylabel="Aversion measure r(w)")
end
```

A linear utility function gives rise to only risk neutral behavior under the Arrow-Pratt model. To see this, let's compute the risk aversion measure $r(w)$. To see this, let's compute the risk aversion measure $r(w)$. The first and second derivatives of the utility function are given by:

## Power Utility function $U(w) = w^{\tau}$
A power utility function gives rise to all three types of choice behavior, under the Arrow-Pratt model, depending upon the choice of the risk aversion parameter. To see this, let's compute the risk aversion measure $r(w, \tau)$ for different values of $\tau$. The first and second derivatives of the utility function are given by:

$$
\begin{eqnarray}
U^{\prime}(w) & = & \tau\cdot{w}^{\tau-1} \\
U^{\prime\prime}(w) & = & (\tau - 1)\cdot\tau\cdot{w}^{\tau-2} \\
\end{eqnarray}
$$

which gives the risk aversion measure $r(w, \tau)$ of

$$
\begin{equation}
r(w,\tau) = - \frac{(\tau - 1)\cdot{w}^{\tau-2}}{w^{\tau-1}}
\end{equation}
$$

which is can on different behaviors depending upon the value of the parameter $\tau$.

### Simulation $\tau < 1$

```julia
let
    
     = 0.1;
    R(w,) = - (( - 1)*w^( - 2))/(w^( - 1));
    
    simulation_array = Array{Float64,2}(undef, number_of_steps, 2)
    for i  eachindex(W)
        simulation_array[i,1] = W[i]
        simulation_array[i,2] = R(W[i],)
    end
    
    plot(simulation_array[:,1], simulation_array[:,2], lw=3, label="U(w) = w^:  = $()", c=:navyblue,
        xlabel="Wealth w", ylabel="Aversion measure r(w)")
end
```
